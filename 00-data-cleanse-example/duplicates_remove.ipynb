```python
from __future__ import print_function
from datetime import datetime
import time
import pandas as pd
import csv
from pathlib import Path
```

```python
print("Start time: " + str(datetime.now()))
```

```python
# Define files
file_in = "Resources/Police_Department_Incident_Reports__2018_to_Present.csv"
file_out ='Resources/SFPD_clean.csv'

df= pd.read_csv(file_in)
df.head()
```

```python
# Check if duplicate file is found before running to make clean file
file_out_dupes = Path('Resources/SFPD_duplicates.csv')
if file_out_dupes.is_file():
    print("Duplicates found")
    # CSV module to specify input and out files
    df.drop_duplicates(subset='Incident Number', keep='last', inplace=True)
else:
    pass
```
